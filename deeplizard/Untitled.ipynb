## Over Fiting

اور فیلترینگ زمانی رخ می‌دهد که دیتای ترین نتایج خیلی خوبی می‌دهد و دیتای تست نتایج خیلی بدی

همینطور برای دیتای ولیدیشن و ترین

برای کاهش اورفیتینگ اگر می‌توانم باید داده های بیشتری را ترین کنیم

روش دیگر برای درست کردن اور فیتینگ

Data augmentation 

است

به این معنی است که دیتا را خودمان تغییر میدهیم

مثلا برای عکس سگ می‌توانم آن را روتین کنیم یا برعکس کنیم

روش دیگر برای از بین بردن اورفیتینگ این است که از پیچیدگی دوری کنیم

مثلا تعداد نورون ها را کم کنیم یا تعداد لایه ها را کاهش دهیم

روش دیگر 

Dropout 

است

این روش اگر به مدل اضافه شود بعضی از ساب ست ها را در نود ها و داده ها ایگنور می‌کند و یک سری نویز ها یا اوتلایر ها را میگیرد

## Under fitting


زمانی رخ می‌دهد که داده ی ترین خوب ترین نمی‌شود و لاس فانکشن بالا می‌باشد مثل ۸.۶۷۸ یا یک چنین چیزی ولی اکیورسی پایین مثل ۰.۴۵ یا چنین چیزی و خوب پردیکت نمیکند

برای از بین بردن آندر فیتینگ 
پیچیدگی را افزایش میدهیم
مثلا تعداد لایه ها را افزایش میدهیم
تعداد نورون ها در هر لایه را افزایش دهیم
یا تایپ لایه ها را عوض کنیم
روش دیگر اضافه کردن فیوچر یا ویژگی های بیشتر به نمونه های ورودی در داده های ترینینگ است
مثلا کالومن های بیشتر اضافه کنیم یا جزییات بیشتر را وارد کنیم

روش دیگر کاهش 

dropout

است

اگر از این روش استفاده میکنیم درصد دراپ اوت را کاهش دهیم
مثلا اگر ۳۰٪ کاهش داریم آن را ۱۵٪ کنیم
